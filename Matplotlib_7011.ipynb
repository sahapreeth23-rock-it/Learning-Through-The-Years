```python
import matplotlib.pyplot as plt
X=[2,4,6,8,10]
Y=[22,24,26,28,30]
plt.figure(figsize=(8,5))
plt.plot(X,Y,marker='o',markersize=10,linestyle='--',color='red',linewidth=2,
         label='sales_data')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly sales report")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
X1=[1,2,3,4,5]
Y1=[1,2,3,4,5]
X2=[5,4,3,2,1]
Y2=[5,4,3,2,1]

fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].plot(X1,Y1,marker='o')
ax[0].set_title("square numbers")
ax[0].set_xlabel("X values")
ax[0].set_ylabel("Y values")

ax[1].plot(X2,Y2,marker='o')
ax[1].set_title("square numbers")
ax[1].set_xlabel("X values")
ax[1].set_ylabel("Y values")

plt.tight_layout()
plt.show()
```
